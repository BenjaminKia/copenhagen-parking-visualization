```python
import pandas as pd
```

```python
df = pd.read_csv("../data/p_pladser.csv")
```

```python
df.shape
```

```python
df["wkb_geometry"] = df["wkb_geometry"].astype(str)
```

```python
# Testing how to get lat
df["wkb_geometry"].str.split(" ").str.get(1).str.replace("(", "")
```

```python
# Create lat column
df["lng"] = df["wkb_geometry"].str.split(" ").str.get(1).str.replace("(", "")
```

```python
# Testing how to get lng values cleaned
df["wkb_geometry"].str.split(" ").str.get(2).str.replace(",", " ")
```

```python
# Create lng col
df["lat"] = df["wkb_geometry"].str.split(" ").str.get(2).str.replace(",", " ")
```

```python
df["lat"]
```

```python
df["lng"]
```

```python
df["aar_mnd"] = df["aar_mnd"].astype(str)
df["year"] = df["aar_mnd"].str[:4]
df["month"] = df["aar_mnd"].str[4:]
```

```python
df["year"]
```

```python
df["month"]
```

```python
df["parkerede_biler_kl_12"] = pd.to_numeric(
    df["parkerede_biler_kl_12"], errors="coerce"
)
df["parkerede_biler_kl_12"] = df["parkerede_biler_kl_12"].fillna(0)
df["parkerede_biler_kl_12"] = df["parkerede_biler_kl_12"].clip(lower=0)
```

```python
df
```

```python
df = df.dropna(subset=["lat", "lng"])
```

```python
df.to_csv("../data/cleaned_df.csv")
```
